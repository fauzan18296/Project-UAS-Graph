```python
import networkx as nx
from pyvis.network import Network
```

```python
# Bangun graf
G = nx.Graph()

G.add_edge("A", "B", weight=1)
G.add_edge("B", "C", weight=2)
G.add_edge("A", "C", weight=3)
G.add_edge("C", "D", weight=4)

path = nx.floyd_warshall(G,weight="weight")
print("Jalur terpendek dari A ke D:", path.get("A").get("D"))

for u, v, data in G.edges(data=True):
    data["label"] = str(data["weight"])

file_generate_graf = "graph.html"
net = Network()
net.from_nx(G)
net.save_graph(file_generate_graf)
print(f"Graf telah disimpan sebagai '{file_generate_graf}'")
```
