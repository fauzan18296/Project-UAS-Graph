```python
import networkx as nx
from pyvis.network import Network
```

```python
def get_path(pred, start, end):

    path = [end]

    while end != start:
        end = pred[start][end]
        path.append(end)
    path.reverse()
    return path
```

```python
# Membuat graf
G = nx.Graph()

# ======================
# NODE (LOKASI)
# ======================

nodes = {
    # Tangga & Koridor
    "v1": "tangga_kiri_lt1",
    "v2": "koridor_kiri_lt1",
    "v3": "koridor_tengah_lt1",
    "v4": "koridor_kanan_lt1",

    # Area bawah
    "v5": "lab_pengembangan_pembelajaran",
    "v6": "lab_distribusi_dan_instalasi_listrik",
    "v7": "lab_konversi_energi",
    "v8": "perpustakaan",
    "v9": "auditorium",

    # Area atas
    "v10": "gudang",
    "v11": "r_hmj",
    "v12": "r_dapur",
    "v13": "workshop_elektro",
    "v14": "krk_elektronika",
    "v15": "krk_elektro",
    "v16": "r_kuliah_s3",

    # Fasilitas
    "v17": "toilet_wanita_kiri",
    "v18": "toilet_pria_kiri",

    "v19": "toilet_wanita_kanan",
    "v20": "toilet_pria_kanan",

    "v21": "plaza",

    # Exit
    "v22": "pintu_keluar_bawah",
    "v23": "pintu_keluar_kanan",

    # Tangga kanan
    "v24": "tangga_kanan_lt1",

    # =====================
    # LANTAI 2
    # =====================

    "v25": "tangga_kiri_tengah_lt2",
    "v26": "koridor_atas_lt2",
    "v27": "koridor_tengah_lt2",
    "v28": "koridor_bawah_lt2",

    "v29": "toilet_pria_lt2",
    "v30": "toilet_wanita_lt2",

    "v31": "r_kuliah_214",
    "v32": "r_kuliah_213",
    "v33": "r_kuliah_212",
    "v34": "r_kuliah_211",

    "v35": "musholla",

    "v36": "lab_digital",
    "v37": "workshop_robotika",
    "v38": "lab_mikroprosesor",

    "v39": "tangga_kanan_lt2",

    "v40": "r_kuliah_205",
    "v41": "r_kuliah_204",
    "v42": "lab_pkdse",
    "v43": "lab_sistem_kendali",
    "v44": "lab_elektronika",
    "v45": "tangga_kiri_lt2"
}

# Tambahkan semua node
for node_id, nama_ruangan in nodes.items():
    G.add_node(node_id)

# ======================
# EDGE (JALUR)
# ======================

edges = [

    # ===== Koridor utama lt1 =====

    ("v1","v2",1),
    ("v2","v3",1),
    ("v3","v4",1),
    ("v4","v24",1),

    # ===== Area bawah =====

    ("v2","v5",1),
    ("v2","v6",1),

    ("v3","v7",1),
    ("v3","v8",1),

    ("v4","v9",1),

    # ===== Area atas =====

    ("v3","v10",1),
    ("v3","v11",1),
    ("v3","v12",1),

    ("v4","v13",1),
    ("v4","v14",1),
    ("v4","v15",1),
    ("v4","v16",1),

    # ===== Toilet kiri =====

    ("v2","v17",1),
    ("v2","v18",1),

    # ===== Toilet kanan =====

    ("v4","v19",1),
    ("v4","v20",1),

    # ===== Plaza =====

    ("v4","v21",1),

    # ===== Exit =====

    ("v3","v22",1),
    ("v21","v23",1),
]

edges += [

    # Tangga antar lantai

    ("v1","v25",2),
    ("v24","v39",2),

    # Koridor

    ("v26","v27",1),  # atas -> tengah
    ("v27","v28",1),  # tengah -> bawah

    # Koridor bawah menuju tangga
    ("v28","v25",1),
    ("v28","v39",1),

    # Area atas

    ("v26","v29",1),
    ("v26","v30",1),

    ("v26","v31",1),
    ("v26","v32",1),
    ("v26","v33",1),
    ("v26","v34",1),
    ("26", "v45",1),

    # Musholla

    ("v27","v35",1),

    # Area tengah

    ("v27","v36",1),
    ("v27","v37",1),
    ("v27","v38",1),

    # Area bawah

    ("v28","v40",1),
    ("v28","v41",1),
    ("v28","v42",1),
    ("v28","v43",1),
    ("v28","v44",1),
]

# Tambahkan edge
for u, v, weight in edges:
    G.add_edge(u, v, weight=weight)

# ======================
# FLOYD WARSHALL
# ======================

pred, dist = nx.floyd_warshall_predecessor_and_distance(
    G,
    weight="weight"
)

asal = "v33"
tujuan = "v23"

path = get_path(pred, asal, tujuan)

print(f"Jalur sebelum menuju tujuan (Predecessor): {pred[asal][tujuan]}")
print(f"Jarak: {dist[asal][tujuan]}")

print("Path:")

for p in path:
    print(nodes[p])

# ======================
# LABEL EDGE
# ======================

for u, v, data in G.edges(data=True):
    # Untuk memunculkan weight nya sebagai label dari edge/sisinya.
    data["label"] = str(data["weight"])

# ======================
# VISUALISASI
# ======================

net = Network(height="800px", width="100%", notebook=True, cdn_resources="remote")

net.from_nx(G)
file_generate_graf = "graph.html"

print("\nGraf disimpan ke ")
net.show(file_generate_graf)
```
